```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Step-2 "Load the Dataset"
 We have three CSV files which we took from the "Kaggle Dataset".
1. train.csv (sales data)
2. features.csv (Economic factors)
3. stores.csv (Store information).

```python
train = pd.read_csv("train.csv")
features = pd.read_csv("features.csv")
stores = pd.read_csv("stores.csv")

print("Train Shape:", train.shape)
print("Features Shape:", features.shape)
print("Stores Shape:", stores.shape)
```

Step-3 "Converting Date column to DataTime Formate for proper time-based analysis".
   

```python
train["Date"] = pd.to_datetime(train["Date"])
features["Date"] = pd.to_datetime(features["Date"])

print("Date Conversion Completed")
```

```python
train.info()
```

##Step-4 "Merging Dataset"
Step-1: Merge train and features using Store and Date.
Step-2: Merge with stores using Store.

```python
# Merge train and features
df = pd.merge(train, features, on=["Store","Date"],how="left")

# Merge with stores
df = pd.merge(df, stores, on="Store", how="left")

print("Final DataSet Shape:", df.shape)
df.head()
```

Step-5: "Check Missing Values"
        Identifying null values before modeling.

```python
missing_values = df.isnull().sum()

print(missing_values)
```

```python
##"As we can see here there are multiple missing values available in MarkDown1 to MarkDown5 so we Fill missing values in MarkDown columns with 0"

markdown_cols = ["MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]
df[markdown_cols] = df[markdown_cols].fillna(0)

print("Missing values handled successfully")
```

```python
# Weekly Sales Distribution

plt.figure(figsize=(8,5))
plt.hist(df["Weekly_Sales"], bins=50)
plt.title("Weekly Sales Distribution")
plt.xlabel("Weekly Sales")
plt.ylabel("Frequency")
plt.xlim(0, 200000)
plt.show()
```

```python
# See sales trend with time ("time se sath sales ka trend dekhna")

sales_trend = df.groupby("Date")["Weekly_Sales"].sum()

plt.figure(figsize=(10,5))
plt.plot(sales_trend)
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Weekly Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# store-wise sales comparision

store_sales = df.groupby("Store")["Weekly_Sales"].sum()

plt.figure(figsize=(10,5))
plt.bar(store_sales.index, store_sales.values)
plt.title("Store-wise Sales Comparison")
plt.xlabel("Store")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Holiday impact on sales

holiday_sales = df.groupby("IsHoliday_x")["Weekly_Sales"].mean()

plt.figure(figsize=(6,5))
plt.bar(holiday_sales.index.astype(str), holiday_sales.values)
plt.title("Holiday vs Non-Holiday Sales")
plt.xlabel("Is Holiday")
plt.ylabel("Average Weekly Sales")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Week"] = df["Date"].dt.isocalendar().week
```

```python
# Checking for Outlires in our dataset

plt.figure(figsize=(8,5))
plt.boxplot(df["Weekly_Sales"])
plt.title("Boxplot of Weekly Sales")
plt.show()
```

```python
Q1 = df["Weekly_Sales"].quantile(0.25)
Q3 = df["Weekly_Sales"].quantile(0.75)
IQR = Q3-Q1

lower = Q1-1.5*IQR
upper = Q3-1.5*IQR

outliers = df[(df["Weekly_Sales"]<lower)|(df["Weekly_Sales"]>upper)]
print("Number of Outliers", len(outliers))
```
